## Importing Modules

Modules are essential in Python as they allow you to organize and reuse code effectively. This section will cover the basics of importing modules, understanding the differences between modules, libraries, and packages, and how to work with Python's Standard Library.

### What is a Module?

A **module** is a single Python file that contains variables, functions, classes, and runnable code. Modules help in organizing code into manageable sections, making it easier to maintain and reuse.


### Modules vs. Libraries vs. Packages

- **Module:** A single Python file containing Python code (e.g., `math.py`).
- **Library:** A collection of related modules providing specific functionalities (e.g., the `math` library).
- **Package:** A directory containing multiple modules and possibly sub-packages, organized in a hierarchical structure (e.g., `numpy` package).


````python
# Visual Presentation
print("""
Library
├── Package
│   ├── Module
│   └── Sub-package
│       └── Module
└── Module
""")
````

### Importing Modules

There are several ways to import modules in Python:

1. **Import the entire module:**




````python
# Import the entire math module
import math

print("The value of pi is", math.pi)
print("The square root of 16 is", math.sqrt(16))
````

2. **Import specific attributes from a module:**


````python
# Import specific functions from math
from math import e, pow

print("The value of e is", e)
print("2 raised to the power of 3 is", pow(2, 3))
````

````python
from string import digits #contains all the digits from '0' to '9'

print("The digits are", digits)
````

3. **Import a module with an alias:**

````python
import math as m

print(m.sqrt(16))
````

**Examples:**

### Third-Party Libraries

In addition to the Standard Library, Python has a vast ecosystem of **third-party libraries** developed by the community. These libraries extend Python's capabilities and are often tailored for specific tasks, such as web development, data analysis, machine learning, and more.

Two popular third-party libraries are `random` and `cowsay`.

#### The `random` Module

The `random` module is part of Python's Standard Library and provides functions for generating random numbers, making random choices, and shuffling sequences.

**Common Functions in `random`:**
- `random.randint(a, b)`: Returns a random integer N such that `a <= N <= b`.
- `random.choice(sequence)`: Returns a randomly selected element from a non-empty sequence.
- `random.shuffle(list)`: Shuffles the elements of a list in place.
- `random.random()`: Returns a random floating-point number in the range `[0.0, 1.0)`.

**Examples:**

````python
# Import the random module
import random

# Generate a random integer between 1 and 10
random_int = random.randint(1, 10)
print("Random integer between 1 and 10:", random_int)

# Select a random element from a list
fruits = ['apple', 'banana', 'cherry', 'date']
random_fruit = random.choice(fruits)
print("Random fruit selected:", random_fruit)

# Shuffle a list
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print("Shuffled numbers:", numbers)

# Generate a random float between 0 and 1
random_float = random.random()
print("Random float between 0 and 1:", random_float)
````

## Calling Functions

Functions are reusable blocks of code that perform specific tasks. Understanding how to call functions is fundamental to writing efficient Python code.

### Function Syntax

To call a function, use the function name followed by parentheses. If the function requires arguments, pass them inside the parentheses.

**Example:**


````python
result = round(3.75)
print(result)  # Output: 4
````

### Functions vs. Programs

- **Program:** A complete, standalone script that performs a series of tasks.
- **Function:** A smaller, reusable piece of code within a program that performs a specific task.

**Diagram:**

### Functions vs. Methods

- **Function:** Defined independently and can be called on their own.
- **Method:** A function that is associated with an object (e.g., strings, lists).

**Example:**


````python
message = "MAKE THIS LOWERCASE"
message2 = message.lower()  # 'lower()' is a method of string objects
print(message2)
````

````python
text = "I AINT YELLIN"
text2 = text.lower()
print(text2)
````

**Question:**
Why does the `lower()` **method** have no arguments while the `round()` **function** requires an argument?

- **Method:** `lower()` is a method of the string object `message`. It knows which string to operate on, so no additional arguments are needed.
- **Function:** `round()` is a standalone function that requires a number as an argument to know which value to round.

### Returning Values

Functions can return values, which can then be used elsewhere in your code.

**Example:**


````python
new_value = round(3.9)
print(new_value)
new_value += 1
print(new_value)
````

````python
new_value = round(19.75)
print(new_value)
new_value = new_value + 10
print(new_value)
````

## Built-in Functions

Python offers a variety of built-in functions that perform common tasks. While it's impossible to cover all of them here, understanding a few key functions will greatly enhance your coding efficiency.

### Common Built-in Functions

- **`abs(x)`**: Returns the absolute value of a number.
- **`bool(x)`**: Converts a value to a Boolean (`True` or `False`).
- **`int(x)`**: Converts a value to an integer.
- **`pow(x, y)`**: Raises `x` to the power of `y`.
- **`len(s)`**: Returns the length of a string, list, tuple, etc.
- **`str(x)`**: Converts a value to a string.
- **`sum(iterable)`**: Sums items of an iterable.
- **And many others...**

**Examples:**


````python
print(abs(-10))           # Output: 10
print(bool(0))            # Output: False
print(int(3.14))          # Output: 3
print(pow(2, 3))          # Output: 8
print(len("Hello"))     # Output: 5
print(str(100))           # Output: "100"
print(sum([1, 2, 3, 4]))  # Output: 10
````

### Nesting Function Calls

A common task is to nest function calls inside of other function calls. When nesting function calls, you read from the inside outward. This can be a little unintuitive, but it may help to treat the parentheses the same you would in a math operation. However, it's critical to remember that the parentheses are there because they are what call the function—not because they are doing math.


T2VBqXQYhl6bQjLfGNkZyjlqViXuQq_KvlENcUMAhxk.png

**Example:**



````python
result = len("hello" + "world".strip())
print(result)  # Output: 10
````



**Explanation:**
1. `"world".strip()` removes any leading/trailing whitespace (though none in this case), resulting in `"world"`.
2. `"hello" + "world"` concatenates to `"helloworld"`.
3. `len("helloworld")` returns `10`.

### Chaining Methods

When you need to call methods on the same string variable or value, it's a little different. As described in the following diagram, you will repeat the period, name, and parentheses each time. By placing them side by side, you will keep using the same variable, modifying it in turn from left to right.

**Example:**


````python
message = "  PYTHON PROGRAMMING  "
clean_message = message.strip().lower().replace("python", "Python")
print(clean_message)
````


**Explanation:**
1. `strip()` removes leading and trailing whitespace.
2. `lower()` converts the string to lowercase.
3. `replace("python", "Python")` replaces "python" with "Python".
